```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d  # Import necessary for Gaussian smoothing
 
# Load the data from a CSV file
data = pd.read_csv('data/experimental/Bubble_shooting.csv')
data.columns = ['X', 'Y']

# Print the first few rows of the data and the existing column names
print("First few rows of data:")
print(data.head())
print("\nCurrent column names:")
print(data.columns)

# Rename the columns assuming there are exactly 2 columns and they are properly ordered
if len(data.columns) == 2:
    data.columns = ['X', 'Y']
    print("\nRenamed column names:")
    print(data.columns)
else:
    print("\nError: Unexpected number of columns. Expected 2 columns but got", len(data.columns))
    exit()  # Exit if the column count is not as expected to prevent further errors

# Proceed if the renaming was successful
if 'X' in data.columns and 'Y' in data.columns:
    # Extract the X and Y columns using the correct column names
    time = data['X']  # Assuming 'X' is the time column
    bubble_area = data['Y']  # Assuming 'Y' is the bubble area column

    # Apply Gaussian smoothing to the bubble area data
    sigma = 0.5  # Adjust the sigma value as needed
    smooth_bubble_area = gaussian_filter1d(bubble_area, sigma)

    # Plot the data
    plt.figure(figsize=(12, 6))  # Set the figure size for better visibility
    plt.plot(time, bubble_area, label='Original Bubble Area', marker='o', linestyle='-', alpha=0.5)
    plt.plot(time, smooth_bubble_area, label='Smoothed Bubble Area', color='red', linewidth=2)

    # Set the x-axis label
    plt.xlabel('Time')

    # Set the y-axis label
    plt.ylabel('Bubble Area')

    # Set the plot title
    plt.title('Bubble Area vs Time')

    # Adding grid for better readability
    plt.grid(True)

    # Show the legend
    plt.legend()

    # Show the plot
    plt.show()

```
